```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

# Define the stopwords and stemmer
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()

def preprocess_text(text):
    # Convert the text to lowercase
    text = text.lower()

    # Remove the punctuation marks
    text = re.sub(r'[^\w\s]', '', text)

    # Tokenize the text
    tokens = nltk.word_tokenize(text)

    # Remove the stopwords
    tokens = [token for token in tokens if token not in stop_words]

    # Perform stemming
    stemmed_tokens = [stemmer.stem(token) for token in tokens]

    # Join the tokens to form the preprocessed text
    preprocessed_text = ' '.join(stemmed_tokens)
    print(preprocessed_text )
#     return preprocessed_text
```

```python
import re

def extract_invoice_number(text):
    invoice_number = re.search(r'(?<=Invoice\sNumber:)\s*\d+', text)
    if invoice_number:
        return invoice_number.group(0)
    return ""

def extract_invoice_date(text):
    invoice_date = re.search(r'\d{1,2}[-/]\d{1,2}[-/]\d{2,4}', text)
    if invoice_date:
        return invoice_date.group(0)
    return ""

def extract_due_date(text):
    due_date = re.search(r'(?<=Due\sDate:)\s*\d{1,2}[-/]\d{1,2}[-/]\d{2,4}', text)
    if due_date:
        return due_date.group(0)
    return ""

def extract_product_line(text):
    product_line = re.search(r'(?<=Product\sline:)\s*[A-Za-z0-9\s]+', text)
    if product_line:
        return product_line.group(0)
    return ""

def extract_unit_price(text):
    unit_price = re.search(r'(?<=Unit\sprice:)\s*\$?\d+(\.\d{2})*', text)
    if unit_price:
        return unit_price.group(0)
    return ""

def extract_quantity(text):
    quantity = re.search(r'(?<=Quantity:)\s*\d+', text)
    if quantity:
        return quantity.group(0)
    return ""

def extract_tax(text):
    tax = re.search(r'(?<=Tax:\s*)\d+(\.\d{2})*\s*%', text)
    if tax:
        return tax.group(0)
    return ""

def extract_total(text):
    total = re.search(r'(?<=Total:\s*)\$\d+(\.\d{2})*', text)
    if total:
        return total.group(0)
    return ""

def extract_date(text):
    date = re.search(r'\d{1,2}[-/]\d{1,2}[-/]\d{2,4}', text)
    if date:
        return date.group(0)
    return ""

def extract_time(text):
    time = re.search(r'\d{1,2}:\d{1,2}(:\d{1,2})?\s*(AM|PM|am|pm)?', text)
    if time:
        return time.group(0)
    return ""

def extract_payment(text):
    payment = re.search(r'(?<=Payment\sMethod:)\s*[A-Za-z0-9\s]+', text)
    if payment:
        return payment.group(0)
    return ""

def extract_telephone(text):
    telephone = re.search(r'(?<=Telephone:\s*)\+?\d+(-\d+)*', text)
    if telephone:
        return telephone.group(0)
    return ""
```
