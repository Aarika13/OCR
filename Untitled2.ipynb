```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
```

```python
#  import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
sent = """Restaurant Receipt

Name Jane Smith 002

Receipt #

Email jasmith@noemail.com  Date 4/8/2019

Phone Number (123) 123-4567 Time 5:00 PM

Ordered Items

Food Name Quantity Price Total

1 Food A 5 5 25

2 Food B 5 5 25

3 Food C 5 5 25

4 Drink D 5 2 10

5 Drink E 5 2 10
Subtotal $95
Tax % 5

I

Total $99.75

Hope you enjoy your dining experience!

1929 Brown Bear Drive, Elsinore,CA, 92330
Phone: (123) 1234567 - nfo@xyzrestaurant com - st xyzrestaurant com"""
words= word_tokenize(sent)
postags=pos_tag(words)
```

```python
ne_tree = nltk.ne_chunk(postags,binary=False)
print(ne_tree)
```

```python
from nltk.chunk import tree2conlltags
from pprint import pprint
iob_tagged = tree2conlltags(ne_tree)
pprint(iob_tagged)
```

```python
import spacy
from pprint import pprint
# Load English tokenizer, tagger, parser and NER
nlp = spacy.load("en_core_web_sm")

# Process whole documents
text = '''Restaurant Receipt

Name Jane Smith 002

Receipt #

Email jasmith@noemail.com  Date 4/8/2019

Phone Number (123) 123-4567 Time 5:00 PM

Ordered Items

Food Name Quantity Price Total

1 Food A 5 5 25

2 Food B 5 5 25

3 Food C 5 5 25

4 Drink D 5 2 10

5 Drink E 5 2 10
Subtotal $95
Tax % 5

I

Total $99.75

Hope you enjoy your dining experience!

1929 Brown Bear Drive, Elsinore,CA, 92330
Phone: (123) 1234567 - nfo@xyzrestaurant com - st xyzrestaurant com

        '''
doc = nlp(text)

# Analyze syntax
print("Noun phrases:", [chunk.text for chunk in doc.noun_chunks])
print("Verbs:", [token.lemma_ for token in doc if token.pos_ == "VERB"])

# Find named entities, phrases and concepts
for entity in doc.ents:
    print(entity.text, entity.label_)
```
