```python
# !pip install keras-ocr

  
import matplotlib.pyplot as plt

import keras_ocr

# keras-ocr will automatically download pretrained
# weights for the detector and recognizer.
pipeline = keras_ocr.pipeline.Pipeline()

# Get a set of three example images
images = [
    keras_ocr.tools.read(url) for url in [
#         'https://upload.wikimedia.org/wikipedia/commons/b/bd/Army_Reserves_Recruitment_Banner_MOD_45156284.jpg',
#         'https://upload.wikimedia.org/wikipedia/commons/e/e8/FseeG2QeLXo.jpg',
#         'https://upload.wikimedia.org/wikipedia/commons/b/b4/EUBanana-500x112.jpg'
        'https://img.freepik.com/free-vector/stylish-modern-poster-design-with-image-space_1017-27359.jpg?w=2000',
        'https://media.geeksforgeeks.org/wp-content/uploads/20210921220033/Annotation20210921215951.jpg'
    ]
]

# Each list of predictions in prediction_groups is a list of
# (word, box) tuples.
prediction_groups = pipeline.recognize(images)

# Plot the predictions
fig, axs = plt.subplots(nrows=len(images), figsize=(20, 20))
for ax, image, predictions in zip(axs, images, prediction_groups):
    keras_ocr.tools.drawAnnotations(image=image, predictions=predictions, ax=ax)
```

Yes, you can use BERT to classify text and create datasets using it. BERT is a pre-trained language model that can be fine-tuned on a specific task such as text classification. To use BERT for text classification, you need to first fine-tune the pre-trained model on a labeled dataset that is similar to the target task. This involves feeding the pre-trained model with the labeled dataset and adjusting the weights of the model through backpropagation.

Once the model is fine-tuned, you can use it to classify new text data. You can also use it to create datasets by labeling new text data and adding it to your existing dataset for fine-tuning the model further.

There are several libraries available in Python such as TensorFlow, PyTorch, and Hugging Face Transformers that provide easy-to-use interfaces for fine-tuning BERT and using it for text classification. These libraries also provide pre-trained BERT models that you can use for various natural language processing tasks including text classification.
